# Pivot Tables

Load the vehicles dataset. We can look at a pivot table where the index is vehicle class:



```python
import pandas as pd
import numpy as np

vehicles = pd.read_csv('data/vehicles.csv')
```

```python
vehicles.head()
```

```python
vehicles.pivot_table(index=["Vehicle Class"])
```

Now, we add the parameter columns:

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=['CO2 Emission Grams/Mile'])
```

Is there any NaN? Why?

```python
# vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=['CO2 Emission Grams/Mile'])
```

Let's use the numpy median function

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG","City MPG"], aggfunc=np.median)
```

Here is an example with combined MPG and CO2 emission grams per mile passed to the values argument.

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG", "CO2 Emission Grams/Mile"])
```

In this example, we will fill the missing values with zero.

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=["Combined MPG"], aggfunc='count', fill_value=0)
```

```python
pd.crosstab(vehicles["Vehicle Class"],vehicles["Cylinders"])
```

## Using cut() to binarize a column

```python
bin_labels_5 = ['Bronze', 'Silver', 'Gold', 'Platinum']
vehicles['levels'] = pd.cut(vehicles['CO2 Emission Grams/Mile'],
                              bins=[0, 300, 500, 700, 1300],
                              labels=bin_labels_5)
```

```python
vehicles.pivot_table(index='Make', columns='levels', values='Year', aggfunc='count', fill_value=0)
```

# Exercise

Build a dataframe with numeric exam marks for the following subjects and find out the student with more A's

- A: 9-10
- B: 7-8
- C: 5-6
- D: 0-4

```python
subjects = ['history', 'english', 'math', 'biology', 'physics', 'chemistry']
```

```python
# create dataframe
notas = pd.DataFrame(np.random.randint(3, 10, (20,6)))
notas.columns = subjects
notas['student'] = notas.index
```

```python
# melt (id_vars, var_name, value_name)
notas = notas.melt(id_vars='student', var_name='subject', value_name='mark')
```

```python
# cut
notas['calification'] = pd.cut(notas.mark, bins=[0,4.9,6.9,8.9,10], labels= ['D','C','B','A'])
```

```python
# pivot index, columns, values, aggfunc, fill_value, sort
notas.pivot_table(index='student',columns='calification', values='subject', aggfunc='count', fill_value=0).sort_values(by='A', ascending=False)
```
